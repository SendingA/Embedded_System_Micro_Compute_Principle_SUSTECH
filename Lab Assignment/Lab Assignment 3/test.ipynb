```python
import cv2
import numpy as np

# 设置图像宽高
width, height = 60, 200

# 创建一个空白图像（渐变背景）
image = np.zeros((height, width, 3), dtype=np.uint8)
for y in range(height):
    color_value = int(255 * (y / height))  # 计算渐变颜色
    image[y, :] = (255, color_value, color_value)  # 渐变的红色背景

# 在图像上绘制文字
text = "Video Player"
font = cv2.FONT_HERSHEY_DUPLEX
font_scale = 0.6
text_thickness = 1
text_color = (0, 0, 0)  # 黑色文字
shadow_color = (180, 180, 180)  # 灰色阴影

# 计算文字大小并居中
(text_width, text_height), _ = cv2.getTextSize(text, font, font_scale, text_thickness)
text_x = (width - text_width) // 2
text_y = (height + text_height) // 2

# 先绘制阴影，再绘制文字
shadow_offset = 2
cv2.putText(image, text, (text_x + shadow_offset, text_y + shadow_offset), font, font_scale, shadow_color, text_thickness)
cv2.putText(image, text, (text_x, text_y), font, font_scale, text_color, text_thickness)

# 增加边框
cv2.rectangle(image, (5, 5), (width - 6, height - 6), (50, 50, 50), 2)  # 灰色边框

# 显示图片
cv2.imshow("Video Player", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 将图像转换为 RGB565 格式
def convert_to_rgb565(image):
    rgb565_data = []
    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            r, g, b = image[y, x]
            rgb565 = ((r & 0xF8) << 8) | ((g & 0xFC) << 3) | (b >> 3)
            rgb565_data.append(rgb565)
    return np.array(rgb565_data, dtype=np.uint16)

# 转换图像
rgb565_image_data = convert_to_rgb565(image)

# 将图像数据保存为数组，方便在C中使用
with open("image_data.txt", "w") as f:
    f.write("{\n")
    for i, color in enumerate(rgb565_image_data):
        f.write(f"0x{color:04X}, ")
        if (i + 1) % width == 0:
            f.write("\n")
    f.write("};\n")

print("图像已转换并保存为 RGB565 格式的数组")
```
